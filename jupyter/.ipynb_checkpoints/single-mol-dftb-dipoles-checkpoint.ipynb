```python
# average IR spectra values
total_amplitude = np.zeros(10000, dtype=np.complex128)
for i in range(0, 128):
    autocorr = ir.autocorrelation(dipoles[i, :, :].T)
    freq, amplitude = ir.ir_spectrum(autocorr, 2e-15)
    total_amplitude += amplitude
```

```python
# normalize amplitude to maximum value
amplitude = np.absolute(amplitude)
amplitude /= np.max(amplitude)
```

```python
# smooth
from scipy.signal import savgol_filter
```

```python
smoothed = savgol_filter(amplitude, 401, 3)
```

```python
plt.plot(freq, smoothed)
plt.xlabel('Frequency (cm^-1)')
plt.ylabel('Normalized absorbance')
plt.xlim([0, 5000])
plt.ylim([0, 0.5])
plt.savefig('../images/single-mol-ir-smoothed.png', dpi=300)
```

```python
# find peaks
from scipy.signal import find_peaks
```

```python
maxima = find_peaks(smoothed[:4000], distance=100)
print(maxima)
```

```python
peaks = [int(freq[i]) for i in maxima[0]]
print(peaks)
```
