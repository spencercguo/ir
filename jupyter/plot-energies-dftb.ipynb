```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open('dftb-h2o-run-clean.ener', mode='r') as f:
    f.readline()
    ke = []
    pe = []
    tot = []
    temps = []
    while(True):
        line = f.readline()
        if (not line): break
        entries = line.split()
        ke.append(float(entries[2]))
        pe.append(float(entries[4]))
        tot.append(float(entries[5]))
        temps.append(float(entries[3]))
```
